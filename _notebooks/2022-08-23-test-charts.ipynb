```python
#hide
import altair as alt
from altair.expr import datum
import logging
import pandas as pd
import sports.yahoo_fantasy.basketball as yahoo_basketball
```

```python
#hide
import logging, sys
logging.disable(sys.maxsize)
basketball = yahoo_basketball.Basketball(league_id="410.l.22889")
```

```python
#hide
draft_df = basketball.draft_results_as_dataframe()
draft_df['price'] = draft_df['cost'].astype(int)
draft_df['team'] = draft_df['team.name'].astype(str)
draft_df['player'] = draft_df['player.name'].astype(str)
```

```python
#hide
def draft_by_round(df: pd.DataFrame, round:int) -> pd.DataFrame:
    d = df[df['round']==round]
    return d[['pick', 'price', 'player', 'team']]

def draft_by_team(df: pd.DataFrame, team:str) -> pd.DataFrame:
    d = df[df['team']==team]
    return d[['round', 'pick', 'price', 'player']]
 
```

In this years competition we had 12 teams and we had 12 roster spots to file, there for we made 144 picks.  

```python
#hide_input
alt.Chart(draft_df, title='Price for each pick').mark_circle().encode(
    x= alt.X("pick", title="Pick"),
    y= alt.Y("price", title="Price [$]"),
    color=alt.Color('team', legend=None),
    tooltip=["round","pick", "price", "player", "team"]
)
```

```python
#hide_input
alt.Chart(draft_df, title='Histogram of prices').mark_bar().encode(
    alt.X('price',bin = alt.BinParams(maxbins = 10), title="Price [$]"),
    y= alt.Y("count()", title="Count"),
)
```

```python
#hide_input
alt.Chart(draft_df).mark_circle().encode(
    x= alt.X("pick", title="Pick", scale=alt.Scale(domain=[0, 144])),
    y= alt.Y("price", title="Price [$]", scale=alt.Scale(domain=[9, 80])),
    color=alt.Color('team', legend=None),
    tooltip=["round","pick", "price", "player", "team"],
    size=alt.Size('round', scale=alt.Scale(domain=[10, 10]), legend=None),
    facet=alt.Facet('team:N', columns=2)
).resolve_axis(
    x='independent',
    y='independent',
)
```

```python
#hide
from IPython.display import display_html
from itertools import chain,cycle
def display_side_by_side(*args,titles=cycle(['']), columns=2):
    html_str='<tr>'
    rows = 0
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
        html_str+='<th style="text-align:center"><td style="vertical-align:top">'
        html_str+=f'<h2 style="text-align: center;">{title}</h2>'
        html_str+=df.to_html(index=False).replace('table','table style="display:inline"')
        html_str+='</td></th>'
        rows+=1
        if rows%columns==0:
            html_str+='</tr><tr>'
    html_str+='<tr>'        
    display_html(html_str,raw=True)
```

```python
#hide_input
display_side_by_side(draft_by_round(draft_df, round=1), draft_by_round(draft_df, round=2), draft_by_round(draft_df, round=3), titles=['A','B','C'], columns=3)
```
